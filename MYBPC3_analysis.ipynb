```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import os
from collections import Counter
import re
from functools import reduce
import scipy.stats as stats
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

def translate_X32(seq):
    """Translate DNA sequence for X32 - G + positions 32:169"""
    table = {
        'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
        'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
        'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
        'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',
        'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
        'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
        'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
        'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
        'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
        'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
        'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
        'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
        'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
        'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
        'TAC':'Y', 'TAT':'Y', 'TAA':'X', 'TAG':'X',
        'TGC':'C', 'TGT':'C', 'TGA':'X', 'TGG':'W',
    }
    protein = ""
    realseq = 'G' + seq[32:169]
    if len(realseq) % 3 == 0:
        for i in range(0, len(realseq), 3):
            codon = realseq[i:i + 3]
            protein += table[codon]
    return protein

def translate_X33(seq):
    """Translate DNA sequence for X33 - positions 30:217 + TG"""
    table = {
        'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
        'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
        'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
        'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',
        'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
        'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
        'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
        'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
        'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
        'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
        'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
        'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
        'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
        'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
        'TAC':'Y', 'TAT':'Y', 'TAA':'X', 'TAG':'X',
        'TGC':'C', 'TGT':'C', 'TGA':'X', 'TGG':'W',
    }
    protein = ""
    realseq = seq[30:217] + 'TG'
    if len(realseq) % 3 == 0:
        for i in range(0, len(realseq), 3):
            codon = realseq[i:i + 3]
            protein += table[codon]
    return protein

def read_fastq_limited(filename, max_bytes):
    """Read FASTQ file up to max_bytes limit."""
    barcodes = []
    bytes_read = 0
    line_count = 0
    
    with open(filename, 'r') as f:
        for line in f:
            bytes_read += len(line.encode('utf-8'))
            if bytes_read > max_bytes:
                print(f"  Reached {max_bytes / (1024**3):.2f} GB limit, stopping read")
                break
            
            if line_count % 4 == 1:
                barcodes.append(line.rstrip())
            line_count += 1
    
    return barcodes

def detect_all_replicates_and_bins(column_names):
    """
    Detect all replicate and bin patterns from column names.
    Handles four patterns:
    1. prefix_r#_b#_*_count
    2. prefix_b#_r#_*_count  
    3. prefix_bin_#_*_count (cardio experiments with underscore)
    4. prefix_bin#_count (cardio experiments without underscore)
    
    Returns dict with (prefix, rep_num) as keys and their bin columns.
    Ignores iPSC and plasmid columns.
    """
    pattern1 = r'(.+?)_r(\d+)_b(\d+)_.+_count'  # r before b
    pattern2 = r'(.+?)_b(\d+)_r(\d+)_.+_count'  # b before r
    pattern3 = r'(.+?)_bin_(\d+)_.+_count'      # prefix_bin_#_*_count
    pattern4 = r'(.+?)_bin(\d+)_count$'         # prefix_bin#_count (no underscore)
    
    replicates = {}  # {(prefix, rep_num): {bin_num: col_name}}
    
    for col in column_names:
        # Skip iPSC and plasmid columns
        if 'iPSC' in col or 'plasmid' in col:
            continue
            
        # Try pattern 1: prefix_r#_b#
        match1 = re.search(pattern1, col)
        if match1:
            prefix = match1.group(1)
            rep_num = match1.group(2)
            bin_num = match1.group(3)
            key = (prefix, rep_num)
            if key not in replicates:
                replicates[key] = {}
            replicates[key][f'b{bin_num}'] = col
            continue
        
        # Try pattern 2: prefix_b#_r#
        match2 = re.search(pattern2, col)
        if match2:
            prefix = match2.group(1)
            bin_num = match2.group(2)
            rep_num = match2.group(3)
            key = (prefix, rep_num)
            if key not in replicates:
                replicates[key] = {}
            replicates[key][f'b{bin_num}'] = col
            continue
        
        # Try pattern 3: prefix_bin_# (cardio experiments with underscore)
        match3 = re.search(pattern3, col)
        if match3:
            prefix = match3.group(1)
            bin_num = match3.group(2)
            # Use prefix as a unique "replicate" identifier for cardio
            key = (prefix, prefix)
            if key not in replicates:
                replicates[key] = {}
            replicates[key][f'b{bin_num}'] = col
            continue
        
        # Try pattern 4: prefix_bin# (cardio experiments without underscore)
        match4 = re.search(pattern4, col)
        if match4:
            prefix = match4.group(1)
            bin_num = match4.group(2)
            key = (prefix, prefix)
            if key not in replicates:
                replicates[key] = {}
            replicates[key][f'b{bin_num}'] = col
            continue
    
    return replicates

def process_exon_data(reads_folder, exon_name, wtseq, splice_regions, translate_func, weights=None, max_file_size_gb=3, exclude_replicates=None):
    """Process sequencing data for a specific exon."""
    
    if weights is None:
        weights = {'b1': 0.1, 'b2': 0.2, 'b3': 0.3, 'b4': 1.0}
    
    if exclude_replicates is None:
        exclude_replicates = []
    
    # Import reads from FASTQ files with size limit
    sequencing_path_list = glob(os.path.join(reads_folder, '*.assembled.fastq'))
    
    max_bytes = max_file_size_gb * 1024 * 1024 * 1024
    
    barcodes_now = {}
    for filename in sequencing_path_list:
        file_size = os.path.getsize(filename)
        print(f"Reading {os.path.basename(filename)} ({file_size / (1024**3):.2f} GB)")
        
        if file_size > max_bytes:
            print(f"  File exceeds {max_file_size_gb} GB, reading first {max_file_size_gb} GB only")
            barcodes = read_fastq_limited(filename, max_bytes)
        else:
            with open(filename, 'r') as f:
                lines = f.readlines()
                barcodes = [line.rstrip() for line in lines[1::4]]
        
        barcodes_now[os.path.basename(filename).split('.')[0]] = barcodes
        print(f"  Loaded {len(barcodes)} reads")
    
    if not barcodes_now:
        raise ValueError(f"No valid FASTQ files found in {reads_folder}")
    
    # Plot total reads
    read_count = [len(s) for s in barcodes_now.values()]
    fig, ax = plt.subplots(figsize=(18.5, 10.5))
    ax.bar(barcodes_now.keys(), read_count, color='g')
    plt.xticks(rotation=90)
    plt.title(f'Total Reads - {exon_name}')
    plt.tight_layout()
    plt.show()
    
    # Calculate unique read frequencies
    barcodes_unique_count = {key: Counter(barcodes_now[key]) for key in barcodes_now.keys()}
    barcodes_freq = {}
    for lib in barcodes_now.keys():
        barcodes_freq[lib] = {key: count/len(barcodes_now[lib]) 
                              for key, count in barcodes_unique_count[lib].items()}
    
    # Plot unique reads
    unique_read_count = [len(set(barcodes_now[key])) for key in barcodes_now.keys()]
    fig, ax = plt.subplots(figsize=(18.5, 10.5))
    ax.bar(barcodes_now.keys(), unique_read_count, color='b')
    plt.xticks(rotation=90)
    plt.title(f'Unique Reads - {exon_name}')
    plt.tight_layout()
    plt.show()
    
    # Plot frequency distributions
    for key in barcodes_now.keys():
        print(key)
        values = list(barcodes_freq[key].values())
        plt.hist(values, histtype='step', log=True, bins=10**np.linspace(-6, 0, 100))
        plt.xscale('log')
        plt.title(key)
        plt.show()
    
    # Build master dataframe
    gate_freq = 1e-5
    
    dataframes = []
    for lib, freq_data in barcodes_freq.items():
        filtered = {seq: freq for seq, freq in freq_data.items() if freq >= gate_freq}
        df = pd.DataFrame(list(filtered.items()), columns=['seq', f'{lib}_count'])
        dataframes.append(df)
    exon_df = reduce(lambda left, right: pd.merge(left, right, on='seq', how='outer'), dataframes)
    
    # Detect all replicates and bins
    all_replicates = detect_all_replicates_and_bins(exon_df.columns)
    
    if not all_replicates:
        print(f"Warning: No scoring columns detected")
        print("Skipping score calculation and normalization")
        
        # Still do filtering and consequence annotation
        aa_wt = translate_func(wtseq)
        wt_len = len(wtseq)
        
        def count_snvs(seq):
            return sum(a != b for a, b in zip(seq, wtseq))
        
        mask = (exon_df['seq'].str.len() == wt_len) & \
               (~exon_df['seq'].str.contains('N')) & \
               (exon_df['seq'].apply(count_snvs) <= 1)
        
        exon_df = exon_df[mask].copy()
        exon_df['type'] = exon_df['seq'].apply(lambda s: 'WT' if s == wtseq else 'SNV')
        
        def annotate_consequence(s):
            if s == wtseq:
                return 'WT'
            
            translation = translate_func(s)
            conseq = sum(1 for a, b in zip(translation, aa_wt) if a != b)
            
            if conseq == 0:
                if s[0:19] != wtseq[0:19] or s[splice_regions['end_junk']:] != wtseq[splice_regions['end_junk']:]:
                    return 'Junk'
                elif s[splice_regions['canon_splice_1'][0]:splice_regions['canon_splice_1'][1]] != \
                     wtseq[splice_regions['canon_splice_1'][0]:splice_regions['canon_splice_1'][1]] or \
                     s[splice_regions['canon_splice_2'][0]:splice_regions['canon_splice_2'][1]] != \
                     wtseq[splice_regions['canon_splice_2'][0]:splice_regions['canon_splice_2'][1]]:
                    return 'Canon_splice'
                elif s[splice_regions['intron_1'][0]:splice_regions['intron_1'][1]] != \
                     wtseq[splice_regions['intron_1'][0]:splice_regions['intron_1'][1]] or \
                     s[splice_regions['intron_2'][0]:splice_regions['intron_2'][1]] != \
                     wtseq[splice_regions['intron_2'][0]:splice_regions['intron_2'][1]]:
                    return 'intron'
                elif any(s[r[0]:r[1]] != wtseq[r[0]:r[1]] for r in splice_regions['splice_regions']):
                    return 'Splice region'
                else:
                    return 'Synonymous'
            
            elif conseq == 1:
                diff_pos = [i for i in range(len(translation)) if translation[i] != aa_wt[i]][0]
                if translation[diff_pos] == 'X':
                    return 'Nonsense'
                elif s[0:19] != wtseq[0:19] or s[splice_regions['end_junk']:] != wtseq[splice_regions['end_junk']:]:
                    return 'Junk'
                else:
                    return 'missense'
        
        exon_df['consequence'] = exon_df['seq'].apply(annotate_consequence)
        
        # Remove Junk
        exon_df = exon_df[exon_df['consequence'] != 'Junk'].copy()
        
        exon_df['average_score_norm'] = np.nan
        
        return exon_df
    
    # USE ALL DETECTED EXPERIMENTS AS REPLICATES
    print(f"\n{'='*80}")
    print("DETECTED EXPERIMENTS (all will be used as replicates):")
    print(f"{'='*80}")
    
    all_experiments = {}  # {unique_rep_id: bin_cols}
    rep_counter = 0
    
    for (prefix, rep_num), bins in all_replicates.items():
        # Create a unique replicate ID
        unique_rep_id = f"{prefix}_rep{rep_num}"
        all_experiments[unique_rep_id] = bins
        print(f"  Replicate {rep_counter + 1}: {unique_rep_id} with bins {sorted(bins.keys())}")
        rep_counter += 1
    
    print(f"\nTotal replicates to use for scoring: {len(all_experiments)}")
    
    # Filter out excluded replicates if any
    if exclude_replicates:
        all_experiments = {r: bins for r, bins in all_experiments.items() if r not in exclude_replicates}
        print(f"After excluding {exclude_replicates}: {len(all_experiments)} replicates")
    
    # Filter sequences
    aa_wt = translate_func(wtseq)
    wt_len = len(wtseq)
    
    def count_snvs(seq):
        return sum(a != b for a, b in zip(seq, wtseq))
    
    mask = (exon_df['seq'].str.len() == wt_len) & \
           (~exon_df['seq'].str.contains('N')) & \
           (exon_df['seq'].apply(count_snvs) <= 1)
    
    exon_df = exon_df[mask].copy()
    exon_df['type'] = exon_df['seq'].apply(lambda s: 'WT' if s == wtseq else 'SNV')
    
    # Annotate consequences
    def annotate_consequence(s):
        if s == wtseq:
            return 'WT'
        
        translation = translate_func(s)
        conseq = sum(1 for a, b in zip(translation, aa_wt) if a != b)
        
        if conseq == 0:
            if s[0:19] != wtseq[0:19] or s[splice_regions['end_junk']:] != wtseq[splice_regions['end_junk']:]:
                return 'Junk'
            elif s[splice_regions['canon_splice_1'][0]:splice_regions['canon_splice_1'][1]] != \
                 wtseq[splice_regions['canon_splice_1'][0]:splice_regions['canon_splice_1'][1]] or \
                 s[splice_regions['canon_splice_2'][0]:splice_regions['canon_splice_2'][1]] != \
                 wtseq[splice_regions['canon_splice_2'][0]:splice_regions['canon_splice_2'][1]]:
                return 'Canon_splice'
            elif s[splice_regions['intron_1'][0]:splice_regions['intron_1'][1]] != \
                 wtseq[splice_regions['intron_1'][0]:splice_regions['intron_1'][1]] or \
                 s[splice_regions['intron_2'][0]:splice_regions['intron_2'][1]] != \
                 wtseq[splice_regions['intron_2'][0]:splice_regions['intron_2'][1]]:
                return 'intron'
            elif any(s[r[0]:r[1]] != wtseq[r[0]:r[1]] for r in splice_regions['splice_regions']):
                return 'Splice region'
            else:
                return 'Synonymous'
        
        elif conseq == 1:
            diff_pos = [i for i in range(len(translation)) if translation[i] != aa_wt[i]][0]
            if translation[diff_pos] == 'X':
                return 'Nonsense'
            elif s[0:19] != wtseq[0:19] or s[splice_regions['end_junk']:] != wtseq[splice_regions['end_junk']:]:
                return 'Junk'
            else:
                return 'missense'
    
    exon_df['consequence'] = exon_df['seq'].apply(annotate_consequence)
    
    # Remove Junk
    exon_df = exon_df[exon_df['consequence'] != 'Junk'].copy()
    
    # Calculate replicate scores for ALL experiments
    score_columns = []
    
    for rep_id, bin_cols in all_experiments.items():
        if not bin_cols:
            print(f"Warning: No bins found for replicate {rep_id}")
            continue
        
        # Check if this is a 3-bin experiment (has b1, b2, b3 but not b4)
        has_b4 = 'b4' in bin_cols
        has_b3 = 'b3' in bin_cols
        
        if has_b3 and not has_b4:
            print(f"Replicate {rep_id}: 3-bin experiment detected, using b4 weight for b3")
        
        weighted_cols = []
        for bin_name, bin_col in bin_cols.items():
            # If 3-bin experiment, use b4 weight for b3
            if bin_name == 'b3' and not has_b4:
                weight = weights['b4']
            else:
                weight = weights[bin_name]
            
            weighted_col = f"{rep_id}_{bin_name}_w_freq"
            exon_df[weighted_col] = exon_df[bin_col].fillna(0) * weight
            weighted_cols.append(weighted_col)
        
        score_col = f"{rep_id}_score"
        numerator = exon_df[weighted_cols].sum(axis=1)
        denominator = exon_df[list(bin_cols.values())].fillna(0).sum(axis=1)
        exon_df[score_col] = numerator / denominator
        # Set to NaN where denominator is 0 (variant not in any bin for this replicate)
        exon_df.loc[denominator == 0, score_col] = np.nan
        score_columns.append(score_col)
    
    print(f"\nCalculated scores for {len(score_columns)} replicates")
    
    if score_columns:
        exon_df['avg_score'] = exon_df[score_columns].mean(axis=1)
        if len(score_columns) > 1:
            exon_df['std_score'] = exon_df[score_columns].std(axis=1)
        
        # Normalize scores
        median_synonymous = exon_df[exon_df['consequence'] == 'Synonymous']['avg_score'].median()
        median_nonsense = exon_df[exon_df['consequence'] == 'Nonsense']['avg_score'].median()
        
        scaling_factor = (0 - 1) / (median_nonsense - median_synonymous)
        offset = 1 - scaling_factor * median_synonymous
        
        exon_df['average_score_norm'] = exon_df['avg_score'] * scaling_factor + offset
        
        print(f"\nNormalization:")
        print(f"  Synonymous median: {median_synonymous:.4f}")
        print(f"  Nonsense median: {median_nonsense:.4f}")
        print(f"  Scaling factor: {scaling_factor:.4f}")
        print(f"  Offset: {offset:.4f}")
    else:
        exon_df['average_score_norm'] = np.nan
    
    return exon_df

def calculate_z_scores_from_synonymous(df, score_column='average_score_norm', consequence_column='consequence'):
    """Calculate z-scores based on synonymous variant distribution."""
    result_df = df.copy()
    
    synonymous_df = df[df[consequence_column] == 'Synonymous']
    synonymous_scores = synonymous_df[score_column].dropna()
    
    mu, std = stats.norm.fit(synonymous_scores)
    result_df['z_score'] = (result_df[score_column] - mu) / std
    
    distribution_params = {
        'mean': mu,
        'std': std,
        'n_synonymous': len(synonymous_scores)
    }
    
    return result_df, distribution_params

def plot_histogram(df, x='average_score_norm', y='consequence', z_score_thresholds=[-3, -2], distribution_params=None, filename='histogram.png'):
    """Plot histogram with consequence categories and z-score thresholds."""
    colors = {
        "Nonsense": "#F3001C",
        "Canon_splice": "#d6621a",
        "intron": "#af79c9",
        "missense": "#00BA74",
        "Splice region": "#fdba00",
        "Synonymous": "#456CCC",
        "WT": "#000000",
        "Junk": "#BFBCBC"
    }
    
    # Filter out WT and Junk
    plot_data = df[~df[y].isin(['WT', 'Junk'])].copy()[[x, y]]
    
    fig, ax = plt.subplots(figsize=(12, 8))
    
    sns.histplot(data=plot_data,
                 x=x,
                 hue=y,
                 ax=ax,
                 bins=35,
                 legend=True,
                 fill=True,
                 palette=colors,
                 edgecolor='black',
                 linewidth=1,
                 multiple="stack")
    
    # Add vertical lines at z-score thresholds converted to assay scores
    if distribution_params:
        mu = distribution_params['mean']
        std = distribution_params['std']
        for z_threshold in z_score_thresholds:
            assay_score = mu + z_threshold * std
            ax.axvline(x=assay_score, color='black', linestyle='--', linewidth=2, alpha=0.7)
            ax.text(assay_score, ax.get_ylim()[1] * 0.95, f'Z={z_threshold}', 
                   rotation=90, verticalalignment='top')
    
    ax.set_xlabel('iPSC-SGE cardio score', fontsize=20)
    ax.set_ylabel('Count', fontsize=20)
    ax.tick_params(axis='x', labelsize=18)
    ax.tick_params(axis='y', labelsize=18)
    ax.set_facecolor('white')
    
    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()

# Define sequences and splice regions for X32
wtseq_X32 = 'GCCCTCTCGGTACCAAGTCCTGTCACCAACAGGCATCACCTATGAGCCACCCAACTATAAGGCCCTGGACTTCTCCGAGGCCCCAAGCTTCACCCAGCCCCTGGTGAACCGCTCGGTCATCGCGGGCTACACTGCTATGCTCTGCTGTGCTGTCCGGGGTAGCCCCAAGGTAGGGAACTTTAGGCGCTGGTCCAGGCCGA'

splice_regions_X32 = {
    'canon_splice_1': (30, 32),
    'canon_splice_2': (169, 171),
    'intron_1': (19, 24),
    'intron_2': (177, 185),
    'splice_regions': [(24, 30), (32, 35), (166, 169), (171, 177)],
    'end_junk': 185
}

# Define sequences and splice regions for X33
wtseq_X33 = 'CTTCCCTCCCTCTCTTTACCTTATTTATAGCCCAAGATTTCCTGGTTCAAGAATGGCCTGGACCTGGGAGAAGACGCCCGCTTCCGCATGTTCAGCAAGCAGGGAGTGTTGACTCTGGAGATTAGAAAGCCCTGCCCCTTTGACGGGGGCATCTATGTCTGCAGGGCCACCAACTTACAGGGCGAGGCACGGTGTGAGTGCCGCCTGGAGGTGCGAGGTGAGGAGCCCTCGGGGCCAGGGCCTGGGAGAT'

splice_regions_X33 = {
    'canon_splice_1': (28, 30),
    'canon_splice_2': (217, 219),
    'intron_1': (19, 22),
    'intron_2': (227, 233),
    'splice_regions': [(22, 28), (30, 33), (214, 217), (219, 227)],
    'end_junk': 233
}

# Process X32 and X33 data
cur_dir = os.getcwd()

print("Processing X32 data...")
X32_df = process_exon_data(
    os.path.join(cur_dir, 'X32_reads'),
    'X32',
    wtseq_X32,
    splice_regions_X32,
    translate_X32,
    max_file_size_gb=3,
    exclude_replicates=[]
)

print("\n" + "="*80)
print("Processing X33 data...")
print("="*80 + "\n")
X33_df = process_exon_data(
    os.path.join(cur_dir, 'X33_reads'),
    'X33',
    wtseq_X33,
    splice_regions_X33,
    translate_X33,
    max_file_size_gb=3,
    exclude_replicates=['MYBPC3_X33_rep2']  # Exclude rep2 from X33
)

# Add exon labels and select subset of columns
X32_subset = X32_df[['seq', 'consequence', 'average_score_norm']].copy()
X32_subset['exon'] = 'X32'

X33_subset = X33_df[['seq', 'consequence', 'average_score_norm']].copy()
X33_subset['exon'] = 'X33'

# Concatenate
C10_data = pd.concat([X32_subset, X33_subset], axis=0, ignore_index=True)

# Reorder columns
C10_data = C10_data[['seq', 'exon', 'consequence', 'average_score_norm']]

# Filter out rows with NaN average_score_norm
C10_data_filtered = C10_data[C10_data['average_score_norm'].notna()].copy()

# Calculate z-scores on the combined data
C10_data_filtered, params = calculate_z_scores_from_synonymous(C10_data_filtered)

# Print summary
print("\n" + "="*80)
print("COMBINED X32 + X33 ANALYSIS")
print("="*80)
print(f"\nTotal variants: {len(C10_data_filtered)}")
print(f"X32 variants: {(C10_data_filtered['exon'] == 'X32').sum()}")
print(f"X33 variants: {(C10_data_filtered['exon'] == 'X33').sum()}")
print(f"\nSynonymous distribution parameters: mean = {params['mean']:.4f}, std = {params['std']:.4f}")
print(f"Number of synonymous variants used: {params['n_synonymous']}")

z_score_summary = C10_data_filtered.groupby('consequence')['z_score'].agg(['count', 'mean', 'std', 'min', 'max'])
print("\nZ-score summary by consequence type:")
print(z_score_summary)

for threshold in [-3, -2, -1]:
    below_threshold = C10_data_filtered[C10_data_filtered['z_score'] < threshold].groupby('consequence').size()
    total_by_consequence = C10_data_filtered.groupby('consequence').size()
    percent_below = (below_threshold / total_by_consequence * 100).fillna(0)
    print(f"\nPercent of variants with Z-score < {threshold}:")
    for cons, pct in percent_below.items():
        print(f"  {cons}: {pct:.1f}%")

# Save to CSV
C10_data_filtered.to_csv('MYBPC3_C10_data.csv', index=False)
print("\nSaved combined data to MYBPC3_C10_data.csv")

# Plot
plot_histogram(C10_data_filtered, x='average_score_norm', y='consequence', 
               z_score_thresholds=[-3, -2], distribution_params=params,
               filename='MYBPC3_C10_conseq.png')
```
